*** Rating Table ***

```python
ratingList = ["U", "PG", "12A", "15", "18"]
```

```python
ID = []
for i in count:
    ID.append(i)
```

```python
movieNameSetList = list(set(movieNameList))
unqiue_rating_define_to_movies = {movie: random.choice(ratingList) for movie in movieNameSetList}
randomlyAssignedRating = [unqiue_rating_define_to_movies[movie] for movie in movieNameList]

Rating_df = pd.DataFrame({
    'Id': ID,
    'Name': movieNameList,
    'Rating': randomlyAssignedRating
})
```

```python
Rating_df.to_csv('/content/sample_data/RatingTableData.csv', index=False)
```
